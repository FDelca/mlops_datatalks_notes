## MLOps Introduction

MLOps stands for Machine Learning Operations. MLOps is a core function of Machine Learning engineering, focused on streamlining the process of taking machine learning models to production, and then maintaining and monitoring them. MLOps is a collaborative function, often comprising data scientists, devops engineers, and IT.
[source](https://databricks.com/glossary/mlops)

## Environment Preparation

1. AWS linux machine (cloud), local linux machine, mac, or wsl -> **Local Ubuntu 22.04 LTS machine**


2. python3.9 installed -> one can install it using this [link](https://linuxize.com/post/how-to-install-python-3-
9-on-ubuntu-20-04/)


3. Install a virtual environment management package:
    - `pipenv`
    - `virtualenv` - This was the one I used - [instructions](https://github.com/LDSSA/batch5-students)


4. Software needed:
    - `docker.io` - `sudo apt install docker.io` - [What is a docker?](https://docs.docker.com/get-started/overview/)
    - Allow your user to use docker without sudo - [link](https://docs.docker.com/engine/install/linux-postinstall/)
    - `docker-compose`, which allows you to manage multiple dockers, steps:
    
        1. `wget https://github.com/docker/compose/releases/download/v2.5.1/docker-compose-linux-x86_64 -O docker-compose`
        
        2. Make it executable: `chmod +x docker-compose`
               
        3. We need to access it from any location of our machine, so open bashrc `nano -bashrc` and add `export PATH="${HOME}/Documents/soft:${PATH}"` to the end of the file.
        
    - Check if `docker` and `docker-compose` are installed - `docker --version`
    - If you  cannot check the version of docker-compose inside your venv then use this [link](https://askubuntu.com/questions/1398344/apt-key-deprecation-warning-when-updating-system) to check the solution / or close and open a new terminal
    
    
5. Now, you are ready to install the python packages you need in your virtualenv (e.g.: `pip install pandas`, etc)

## MLOps maturity model

**`Different levels:`**

**0. Nothing Automated**
- No automation
- All code in jupyter (bad coded)
- It is ok, for PoC levels

**1. Devops, No MLOps (Covered in the engineering part, but not in data science one)**
- Releases are automate
- CI/CD
- Operation metrics


- No experiment tracking
- No Reproducibility
- DS separated from engineers


**2. Automated Training**
- Training Pipeline
- Experiment Tracking
- Model registry
- Low Friction but manual deployment
- DS work with engineers.
- Depend always on a DS expertise to deliver other model


**3. Automated Deployment**
- Easy to deploy model - ML Plataform - by an API call
- A/B test to compare models - in the ML Plataform - **not covered in this course**
- Model Monitoring

**4. Full MLOps Automation**
- Automating training, re-training, and deployment

Further explanation [here](https://docs.microsoft.com/en-us/azure/architecture/example-scenario/mlops/mlops-maturity-model)
